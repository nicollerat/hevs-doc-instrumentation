```python
from scipy.signal import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
filtre=firwin(20, 0.1, window="hamming")
```

```python
filtre
```

```python
M=100
# M*k / M = 10 * pi => k=10*pi*M
time=np.narray(M)
sig=np.ones(M)+time*sin(time
fsig=lfilter(filtre,1,sig)
```

```python
plt.plot(sig)
plt.plot(fsig)
```
