## First computation of the FFT

Generate a signal and compute it's FFT.

```python
import numpy as np
import matplotlib.pyplot as plt

N=32
P=128
n=np.arange(N)
n
```

Example of sinusoidal signal

- What about the sinus argument ?
- Try other frequencies, check for limits
- Try to combine several signals

```python
f=3;g1=0
f2=4; g2=1
f3=6; g3=0.3
mon_signal = g1*np.sin(2*np.pi*f*n/N)+g2*np.sin(2*np.pi*f2*n/N)+np.cos(2*np.pi*f3*n/N)*g3
mon_signal
```

```python
plt.plot(mon_signal)
```

```python
sfft=np.fft.fft(mon_signal)
```

```python
f,ax=plt.subplots(2)
ax[0].plot(sfft.real)
ax[0].set_title("real part")
ax[1].plot(sfft.imag)
ax[1].set_title("imaginary part")
plt.tight_layout()
```
