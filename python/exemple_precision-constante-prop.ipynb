```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
N=500
g0=1.0
n0=0.1 # Erreur constante (0.1 typ)
n1=0.5 # Erreur proportionnelle (0.5 typ)

# Points en x
x = np.linspace(0, 1, N)

# Points au hasard selon les paramtres
y1 = (np.random.rand(len(x))-0.5)*2*n1*x
y = g0*x+(np.random.rand(len(x))-0.5)*2*n0 +y1

# Limite constante/proportionnelle
x1 = n0/n1

# Graphique
fig, ax = plt.subplots(1,1,figsize=(15,10))
ax.plot(x,y,'rx')
ax.plot([0,1],[0,1])
ax.plot([0,x1],[n0,n0+x1],'r')
ax.plot([x1,1],[(1+n1)*x1,1+n1],'r')
ax.plot([0,x1],[-n0,-n0+x1],'r')
ax.plot([x1,1],[(1-n1)*x1,1-n1],'r')
ax.plot([x1,x1],[0,0.4],'b.-')
ax.set_xlabel("mesurande")
ax.set_ylabel("sortie")
ax.set_title("Mesure avec bruit et dfaut de gain et d'offset")
ax.grid(True, which='both')
ax.legend(['mesure','idale'])
seaborn.despine(ax=ax, offset=0)
```
